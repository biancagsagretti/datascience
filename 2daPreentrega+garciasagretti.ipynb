# **1 - Detalles del Proyecto**







¿Cuáles serán las emisiones en el futuro?

## **1.1  Objetivo**

El presente proyecto tiene el objetivo de predecir futuras emisiones de Dióxido de Carbono a través de un modelo de regresión. Al predecir las futuras emisiones y relacionándolas con la temperatura promedio, podremos ver el estadío en el que estaremos en los tiempos venideros y poder tomar medidas frente al cambio climático y desarrollar prácticas de producción y distribución más sustentables.


## **1.2 Contexto Empresarial**

Es conocido por todos a través de los medios, redes sociales, libros, revistas académicas, que los niveles de emisión de gases de efecto invernadero hacia la atmósfera ha ido en aumento constante desde los inicios de la era industrial, con un profundo acrecentamiento en el comienzo de lo que conocemos como Globalización. También es sabido que el clima está cambiando y los fenómenos naturales extremos se ven con más frecuencia en diferentes latitudes.

Has sido contratado por un equipo conformado por científicos climáticos, ingenieros, geofísicos,etc, para poder elaborar un modelo que pueda predecir futuras emisiones de CO2 y poder demostrar mediante visualizaciones si existe relación entre las emisiones y el aumento de temperatura, cómo fue la evolución a lo largo de los años y la distribución en los diversos continentes y actividades.


## **1.3 Problema Comercial**

El problema se refiere a que actualmente no se cuenta con un plan o estrategias para disminuir las emisiones porque se desconocen las tasas de crecimiento y la evolución de las mismas.

## **1.4 Contexto Analítico**

El Proyecto se construirá en base al Dataset *"Agri-food CO2 emission dataset"*, un dataset formado por datos obtenidos de la FAO (Food and Agriculture Organization) y datos del IPCC.

El Dataset demuestra las emisiones de CO2 relacionadas al sector agrícola, ganadero, forestal, producción y transporte de de productos alimentarios en todos los países del mundo a lo largo de los años desde 1990 hasta el 2020.
Se encuentran también datos relacionados a la población, tanto rural como urbana, el total de emisiones y la temperatura promedio por año.
Estos sectores representan más de la mitad de emisiones Dióxido de Carbono a nivel global.



Lo primero que hacemos es preparar los datos. Analizaremos la relación entre las emisiones de CO2 y el aumento de la temperatura. Se analizan las emisiones a través de los años, las emisiones por continente y países que más emiten para ver el impacto en el resultado a nivel global. Para ello se utilizan herramientas de visualización.

Se buscará realizar la predicción de emisiones y los cambios de temperatura a través de modelos de regresión y entender a través de estos modelos la relación entre las variables de emisión y temperatura.

Esperamos que la predicción lograda nos permita estimar las emisiones y a partir de allí poder tomar decisiones con respecto a qué hacer frente al escenario que se plantea.

On-farm Electricity Use: consumo de electricidad en granjas.

Food Packaging: emisiones debido a la producción y desecho de paquetes para productos alimenticios.

Agrifood Systems Waste Disposal: emisiones debido a los residuos del sistema agrícola.

Food Processing: emisiones en el procesado de sustancias alimenticias.

Fertilizers Manufacturing: emisiones en la producción de fertilizantes.

IPPU: emisiones en los procesos industriales y el uso del producto.

Manure applied to Soils: emisiones debidas al abono animal en suelos agrícolas.

Manure left on Pasture: emisiones debidas al abono animal en tierras de pastura.

Manure Management: emisiones en el tratamiento de abono animal.

Fires in organic soils: emisiones debido a los incendios en suelos orgánicos.

Fires in humid tropical forests: emisiones en bosques tropicales.

On-farm energy use: consumo de energía en granjas.

Rural population: población en áreas rurales.

Urban population: población en área urbanas.

Total Population - Male: total de población masculina.

Total Population - Female: total de población femenina.

total_emission: total de emisiones de gases de efecto invernadero de diversas fuentes.

Average Temperature °C: el promedio de aumento de temperatura por año expresado en grados Celsius.

```python


url = "https://beta4.api.climatiq.io/search"
authorization_headers = {"Authorization": f"Bearer: {MY_API_KEY}"}

def api_query():
 #la key del diccionoario de datos es un string,
 #tiene el formato {"nombre_key" : "valor"},
 #no podias acceder al valor porque era una variable
 # {query : "Agri..."}

  queries= {
      "query_2": "Agriculture/Hunting/Forestry/Fishing",
      "query_1": "Organic Products Waste",
            }
  querys_params = []

  for key in list(queries.keys()):
    #el return estaba dentro del ambito del for,
    #lo que cortaba el for y no terminaba de armar los queries
    query = {
            "query" : queries[key],
            "data_version": "^1",
                  }
    querys_params.append(query)

  return querys_params

querys_param = api_query()
querys_param
```

```python
texto = requests.get(url, params=querys_param[1], headers=authorization_headers)
texto= texto.text
jsondata = json.loads(texto)
```

```python
#variable para concatenar los resultados
import time
#importo time para que haya un segundo de espera por cada consulta
#esto es para evitar superar las cuotas de las apis

#asigno una variable con un dataframe vacio para los resultados
df_api = pd.DataFrame()

#recorremos el array con los params para la consulta
for query in querys_param:
  texto = requests.get(url, params=query, headers=authorization_headers)
  texto= texto.text
  jsondata=json.loads(texto)
  #como los dataset tienen las mismas columnas concateno los resultados uno debajo de otro
  #concateno lo que tenia en df_api con el resultado nuevo y lo asigno a la variable

  df_api= pd.concat([pd.DataFrame(jsondata["results"]),df_api], ignore_index=True)
  time.sleep(1)
#por ultimo se resetea el index.
df_api = df_api.reset_index(drop=True)
```

```python

def forward_selection(data, target, significance_level=0.01):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

```python
forward_selection(X,y, 0.01)
```

```python
def backward_elimination(data, target,significance_level = 0.05):
    features = data.columns.tolist()
    while(len(features)>0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
        else:
            break
    return features
```

```python
backward_elimination(X,y)
```

```python
def stepwise_selection(data, target,SL_in=0.05,SL_out = 0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<SL_in):
            best_features.append(new_pval.idxmin())
            while(len(best_features)>0):
                best_features_with_constant = sm.add_constant(data[best_features])
                p_values = sm.OLS(target, best_features_with_constant).fit().pvalues[1:]
                max_p_value = p_values.max()
                if(max_p_value >= SL_out):
                    excluded_feature = p_values.idxmax()
                    best_features.remove(excluded_feature)
                else:
                    break
        else:
            break
    return best_features
```
